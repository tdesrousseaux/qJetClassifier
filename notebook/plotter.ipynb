```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
```

```python
ea = []
train_loss = []
train_accuracy = []
val_loss = []
val_accuracy = []

for i in range(4):
    path = f'/data/tdesrous/qJetClassifier/logs/lightning_logs/version_{i}/'
    ea.append(event_accumulator.EventAccumulator(path))
    ea[i].Reload()

    train_loss.append(ea[i].Scalars('train_loss_epoch'))
    train_accuracy.append(ea[i].Scalars('train_accuracy_epoch'))
    val_loss.append(ea[i].Scalars('val_loss_epoch'))
    val_accuracy.append(ea[i].Scalars('val_accuracy_epoch'))


```

```python
steps = []
train_loss_values=[]
train_accuracy_values = []
val_loss_values = []
val_accuracy_values = []

for i in range(4):
    steps.append([x.step for x in train_loss[i]])
    train_loss_values.append([x.value for x in train_loss[i]])
    train_accuracy_values.append([x.value for x in train_accuracy[i]])
    val_loss_values.append([x.value for x in val_loss[i]])
    val_accuracy_values.append([x.value for x in val_accuracy[i]])

```

```python
legend = [r'$S=10000, N_R=3$', r'$S=10000, N_R=5$', r'$S=1000, N_R=3$', r'$S=1000, N_R=5$']
```

```python
for i in range(4):
    plt.plot(steps[i], train_loss_values[i], label=legend[i])

plt.xlabel('steps')
plt.title('Train loss')
plt.legend()
plt.grid()
```

```python
for i in range(4):
    plt.plot(steps[i], val_loss_values[i], label=legend[i])

plt.xlabel('steps')
plt.title('Validation loss')
plt.legend()
plt.grid()
```

```python
for i in range(4):
    plt.plot(steps[i], train_accuracy_values[i], label=legend[i])

plt.xlabel('steps')
plt.title('Train accuracy')
plt.legend()
plt.grid()
```

```python
for i in range(4):
    plt.plot(steps[i], val_accuracy_values[i], label=legend[i])

plt.xlabel('steps') 
plt.title('Validation accuracy')
plt.legend()
plt.grid()
```
